```python
nycviolation_df=pd.read_csv(r'C:\Users\User\OneDrive\Desktop\practice files\violations.csv')
nycviolation_df
```

```python
#cleaning up data
nycviolation_df.replace('BLANKPLATE',np.nan)
nycviolation_df = nycviolation_df[nycviolation_df['Registration State'] != '99']
```

```python
nycviolation_df=nycviolation_df.dropna(subset=['Plate ID','Registration State','Vehicle Make','Street Name'])
nycviolation_df
```

```python
#finding which violation code has been most violeted per state

nycviolation_counts=nycviolation_df.groupby(['Registration State', 'Violation Code'])['Summons Number'].count().reset_index()
nycviolation_counts
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plot with bar graph code per new york 
filt = nycviolation_counts['Registration State'] == 'NY'
newyorkcarsvio = nycviolation_counts[filt]
newyorkcarsvio
```

```python
#for top five violation codes in new york
newyorkcarsvio = nycviolation_counts.loc[nycviolation_counts['Registration State'] == 'NY'].nlargest(5, 'Summons Number')
plt.bar(newyorkcarsvio['Violation Code'],newyorkcarsvio['Summons Number']);
plt.xlabel('Violation Code');
plt.ylabel('Summons Number');
plt.title('Violation Trends for Newyork By Codes');

```

```python
#summone per state
summonses_per_state = nycviolation_df.groupby('Registration State')['Summons Number'].sum().reset_index()
summonses_per_state
```

```python
top_five_states = summonses_per_state.sort_values(by='Summons Number', ascending=False).head(5).reset_index()


top_five_states
```

```python
plt.pie(top_five_states['Summons Number'],labels=top_five_states['Registration State'], autopct='%1.1f%%');
plt.title('Top Five States by Number of Summonses Issued')
plt.show()
```

```python
vehicle_make = nycviolation_df.groupby(['Vehicle Make', 'Registration State'])['Summons Number'].sum().reset_index().sort_values(by='Summons Number',ascending=False).head(10)
vehicle_make
```

```python
plt.plot(vehicle_make['Vehicle Make'],vehicle_make['Summons Number']);
plt.xlabel('Vehicle Make');
plt.ylabel('Summons Number');
plt.title('Summons Trends for Vehiclemake By states');

```

```python
# Pivot the DataFrame to create a matrix of counts for each combination of state and violation code
nycviolation_pivot = nycviolation_counts.pivot(index='Registration State', columns='Violation Code', values='Summons Number').head(30)

# Create a heat map with the counts as the color scale
plt.figure(figsize=(12,8))
sns.heatmap(nycviolation_pivot, cmap='Reds', annot=True, fmt='.0f', cbar_kws={'label': 'Number of Violations'})
plt.title('Violations by State and Violation Code')
plt.xlabel('Violation Code')
plt.ylabel('Registration State')
plt.show()
```
